```python

import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [10, 20, 30, 40, 50, 60]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Group by 'Category' column
grouped = df.groupby('Category')

# Sum of values for each category
sum_values = grouped['Values'].sum()
print("\nSum of values for each category:\n", sum_values)

# Mean of values for each category
mean_values = grouped['Values'].mean()
print("\nMean of values for each category:\n", mean_values)

# Count of values for each category
count_values = grouped['Values'].count()
print("\nCount of values for each category:\n", count_values)
```

```python
days  = [ 1,2,3,4,5,6]
tem = [11,12,13,14,15,66]
min =  [22,23,2,44,45,46]
max= [51,62,63,67,68,70]
plt.xlabel("days")
plt.ylabel("tempeartue")
plt.title("weather")
plt.plot(days,min,label='Max')
plt.plot(days,max,label='Min')    
plt.plot(days,tem,label ='Avg')
plt.legend()
```

```python
#we can move the legend max, min,avg,#see the documentation
days  = [ 1,2,3,4,5,6]
tem = [11,12,13,14,15,66]
min =  [22,23,2,44,45,46]
max= [51,62,63,67,68,70]
plt.xlabel("days")
plt.ylabel("tempeartue")
plt.title("weather")
plt.plot(days,min,label='Max')
plt.plot(days,max,label='Min')    
plt.plot(days,tem,label ='Avg')
plt.legend(loc="upper right",fontsize="small") # font size is small, we can move the legend mean  we can locate the upper right , upper left
plt.grid()
```
